```python
#import as funçoes que seram usadas
import matplotlib.pyplot as plt
import geopandas as gpd
import os
```

```python
#lendo o arquivo que usei como exemplo
mun_ba = gpd.read_file("/content/drive/MyDrive/sig/ba_mun1.shp")
```

```python
#lendo a tabela de atributos
mun_ba.head()
```

```python
#lendo um atributo da tabela e declarando ele 
itaparica=(mun_ba[mun_ba['NM_MUN']=='Itaparica'])
```

```python
#salvando essa declaração que fiz para salvar no drive
itaparica.to_file('/content/drive/MyDrive/sig/itaparica.shp')
```

```python
#visualizando o estado da bahia e o novo arquivo itaparica. atribuindo cores e tamanhos. 
estado = mun_ba.plot(color='grey', edgecolor = 'black', linewidth=0.5, figsize=(15,10 ) )
itaparica.plot(ax=estado, color='red')
```
